```python
with open('day5_input') as f:
    lines = f.readlines()
lines = [l.strip() for l in lines]
```

```python
import numpy as np
import re
```

```python
def points(coords):
    start, end = re.split(' -> ', coords)
    startx, starty = re.split(',', start)
    endx, endy = re.split(',', end)
    points = [startx, endx, starty, endy]
    points = [int(point) for point in points]
    return points  # returns (x1, x2, y1, y2)
```

```python
all_points = 0
for line in lines:
    coords = points(line)
    if max(coords) > all_points:
        all_points = max(coords)
```

```python
def straight_line(points):
    x1, x2, y1, y2 = points
    if x1 == x2:
        if y2 > y1:
            return (1, x1, [y1, y2])
        else:
            return (1, x1, [y2, y1])
    elif y1 == y2:
        if x2 > x1:
            return (2, [x1, x2], y1)
        else:
            return (2, [x2, x1], y1)
    else:
        return (0)
```

```python
vent_map = np.zeros((all_points + 1, all_points + 1))
for line in lines:
    coords = points(line)
    straight_data = straight_line(coords)
    print(straight_data)
    if straight_data != 0:  # not straight line
        if straight_data[0] == 1:  # straight in x
            for y in range(straight_data[2][0], straight_data[2][1] + 1):
                print(y)
                vent_map[straight_data[1]][y] += 1
        else:  # straight in y
            for x in range(straight_data[1][0], straight_data[1][1] + 1):
                vent_map[x][straight_data[2]]
```
